```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image = Image.open("New_Zealand_Boat.jpg")
```

```python
plt.imshow(image)
plt.show()
```

```python
def show_image(img, title=""):
    plt.imshow(img)
    plt.title(title)  # Now it accepts a title argument
    plt.axis("off")
    plt.show()
```

```python
# Define augmentation transformations
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),  # Flip image horizontally
    transforms.RandomRotation(degrees=45),   # Rotate randomly between -45° and 45°
    transforms.RandomResizedCrop(size=(224, 224), scale=(0.7, 1.0)),  # Random crop & resize
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.1),  # Color variations
    transforms.RandomGrayscale(p=0.2),  # Convert to grayscale with 20% probability
    transforms.ToTensor()  # Convert image to tensor
])
```

```python
# Generate and save 10 augmented images
for i in range(10):
    augmented_image = transform(image)  # Apply transformation
    augmented_image_pil = transforms.ToPILImage()(augmented_image)  # Convert to PIL image
    show_image(augmented_image_pil, f"Augmented Image {i+1}")  # Display the image
    augmented_image_pil.save(f"augmented_image_{i+1}.jpg")  # Save image
```
